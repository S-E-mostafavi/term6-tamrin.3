```python
# from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ูพุงุฑุงูุชุฑูุง
IMG_HEIGHT = 150
IMG_WIDTH = 150
BATCH_SIZE = 32
VALIDATION_SPLIT = 0.2

# ุงุฌุงุฏ ImageDataGenerator ุจุฑุง augmentation ู normalization
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=VALIDATION_SPLIT
)

# ุชููุฏ ุฌุฑุงู ุฏุงุฏู ุขููุฒุด
train_generator = datagen.flow_from_directory(
    BASE_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

# ุชููุฏ ุฌุฑุงู ุฏุงุฏู validation
validation_generator = datagen.flow_from_directory(
    BASE_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

# ุจุฑุฑุณ ฺฉูุงุณโูุง
print("ฺฉูุงุณโูุง:", train_generator.class_indices)
```

```python
# ุงุฌุงุฏ ูุฏู
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(len(class_names), activation='softmax')
])

# ฺฉุงููพุงู ูุฏู
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# ููุงุด ุฎูุงุตู ูุฏู
model.summary()

# ุขููุฒุด ูุฏู
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE
)
```

```python
# ุชุนุฑู ฺฉ ูุฏู ุนุตุจ ุณุงุฏู
model = keras.Sequential([
    layers.Dense(units=64, activation='relu', input_shape=(input_dimension,)),
    layers.Dense(units=32, activation='relu'),
    layers.Dense(units=num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("ูุฏู ุนุตุจ ุณุงุฏู ุชุนุฑู ุดุฏ:")
model.summary()
print("\n" + "="*50 + "\n")

#  ุขููุฒุด ูุฏู ุจุง ุงุณุชูุงุฏู ุงุฒ ุจฺ
print("ุดุฑูุน ุขููุฒุด ูุฏู ุจุง ุงุณุชูุงุฏู ุงุฒ Batch Size...")

batch_size = 4
epochs = 2

history = model.fit(X_data, y_labels_true,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1)

print(f"\nุขููุฒุด ุจู ูพุงุงู ุฑุณุฏ. (ูุฑ Epoch ุดุงูู {np.ceil(num_samples / batch_size).astype(int)} ฺฏุงู/ุจฺ ุจูุฏ.)")
print("ุฏุฑ ุญู ุขููุฒุดุ ูุฏู ุฏุงุฏูโูุง ุฑุง ุจู ุจฺโูุง 4 ุชุง ุฏุฑุงูุช ู ูพุฑุฏุงุฒุด ฺฉุฑุฏ.")
print("\n" + "="*50 + "\n")
```

```python
# TODO : ฺฉุด ฺฉุฑุฏู ุฏุงุฏูโูุง ุฏุฑ ุญุงูุธู ุจุฑุง ุงูุฒุงุด ุณุฑุนุช ุขููุฒุด (ููุท ุฏุฑ ุตูุฑุช ฺฉุงู ุจูุฏู ุญุงูุธู RAM)
import tensorflow as tf
import numpy as np
import time

num_samples = 10000
input_dim = 100 # ุงุจุนุงุฏ ูุฑ ููููู
num_classes = 10 # ุชุนุฏุงุฏ ฺฉูุงุณโูุง

X_data_simulated = np.random.rand(num_samples, input_dim).astype(np.float32)
y_labels_simulated = np.random.randint(0, num_classes, size=num_samples)

print(f"ุดฺฉู ุฏุงุฏูโูุง ุดุจูโุณุงุฒ ุดุฏู (X_data_simulated): {X_data_simulated.shape}")
print(f"ุดฺฉู ูุจูโูุง ุดุจูโุณุงุฒ ุดุฏู (y_labels_simulated): {y_labels_simulated.shape}")
print("\n" + "="*50 + "\n")

def load_data_slowly(feature, label):

    time.sleep(0.0001) # ุดุจูโุณุงุฒ ุชุงุฎุฑ 0.1 ููโุซุงููโุง ุจุฑุง ูุฑ ููููู
    return feature, label

# ุงุฌุงุฏ tf.data.Dataset ุงุฒ ุฏุงุฏูโูุง NumPy
dataset_original = tf.data.Dataset.from_tensor_slices((X_data_simulated, y_labels_simulated))

# ูฺฏุงุดุช ุชุงุจุน load_data_slowly ุจู ูุฑ ููููู
dataset_processed = dataset_original.map(load_data_slowly)

# ุดุงูู ฺฉุฑุฏู ู ุจฺ ฺฉุฑุฏู
batch_size = 32
dataset_batched = dataset_processed.shuffle(buffer_size=1000).batch(batch_size)

print("ูพุงูพโูุงู ุฏุงุฏู ุจุฏูู ฺฉุด ุงุฌุงุฏ ุดุฏ.")
print("\n" + "="*50 + "\n")

# 3. ุงูุฏุงุฒูโฺฏุฑ ุฒูุงู ุขููุฒุด (ุจุฏูู ฺฉุด)
print("ุดุฑูุน ุขููุฒุด ุจุฏูู ฺฉุด...")
start_time_no_cache = time.time()

epochs = 3

# ุดุจูโุณุงุฒ ุญููู ุขููุฒุด
for epoch in range(epochs):
    print(f"Epoch {epoch+1} (ุจุฏูู ฺฉุด):")
    for batch_features, batch_labels in dataset_batched:

        _ = 1
    print(f"  Epoch {epoch+1} ฺฉุงูู ุดุฏ.")

end_time_no_cache = time.time()
time_taken_no_cache = end_time_no_cache - start_time_no_cache
print(f"\nุฒูุงู ฺฉู ุขููุฒุด ุจุฏูู ฺฉุด: {time_taken_no_cache:.2f} ุซุงูู")

print("\n" + "="*50 + "\n")

dataset_cached = dataset_original.map(load_data_slowly).cache()

# ุดุงูู ฺฉุฑุฏู ู ุจฺ ฺฉุฑุฏู
dataset_cached_batched = dataset_cached.shuffle(buffer_size=1000).batch(batch_size)

print("ูพุงูพโูุงู ุฏุงุฏู ุจุง ูุงุจูุช ฺฉุด ุงุฌุงุฏ ุดุฏ.")
print("\n" + "="*50 + "\n")

print("ุดุฑูุน ุขููุฒุด ุจุง ฺฉุด...")
start_time_cache = time.time()

# ุดุจูโุณุงุฒ ุญููู ุขููุฒุด
for epoch in range(epochs):
    print(f"Epoch {epoch+1} (ุจุง ฺฉุด):")
    for batch_features, batch_labels in dataset_cached_batched:
        _ = 1
    print(f"  Epoch {epoch+1} ฺฉุงูู ุดุฏ.")

end_time_cache = time.time()
time_taken_cache = end_time_cache - start_time_cache
print(f"\nุฒูุงู ฺฉู ุขููุฒุด ุจุง ฺฉุด: {time_taken_cache:.2f} ุซุงูู")

print("\n" + "="*50 + "\n")

#  ููุงุณู ูุชุงุฌ
print("--- ููุงุณู ูุชุงุฌ ---")
print(f"ุฒูุงู ุจุฏูู ฺฉุด: {time_taken_no_cache:.2f} ุซุงูู")
print(f"ุฒูุงู ุจุง ฺฉุด: {time_taken_cache:.2f} ุซุงูู")
if time_taken_no_cache > time_taken_cache:
    print(f"ฺฉุด ฺฉุฑุฏู ุจุงุนุซ ุณุฑุนุช ุจุฎุดุฏู ุจู ุขููุฒุด ุดุฏ! (ุญุฏูุฏ {time_taken_no_cache / time_taken_cache:.2f} ุจุฑุงุจุฑ ุณุฑุนุชุฑ)")
else:
    print("ฺฉุด ฺฉุฑุฏู ุฏุฑ ุงู ูุซุงู ุจูุจูุฏ ูุงุจู ุชูุฌู ุงุฌุงุฏ ูฺฉุฑุฏ (ููฺฉู ุงุณุช ุจู ุฏูู ุฏุชุงุณุช ฺฉูฺฺฉ ุง ุณุฑุนุช ุจุงูุง I/O ุดุจูโุณุงุฒ ุดุฏู ุจุงุดุฏ).")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# ุชุนุฑู ูุงู ุงูุฒุงุด ุฏุงุฏู ุณูุงุฑุด ุงุตูุงุญ ุดุฏู
class RandomRotateFlip(layers.Layer):
    def __init__(self, rotation_factor=0.1, **kwargs):
        super().__init__(**kwargs)


        self.rotation_factor = rotation_factor

        # ุงุฌุงุฏ ูุงู RandomRotation ุฏุฑ ุฒูุงู ุณุงุฎุช ูุงู ุณูุงุฑุด
        self.rotation_layer = layers.RandomRotation(
            factor=rotation_factor,
            seed=None
        )

    def call(self, images, training=True):
        if not training:
            return images

        augmented_images = images

        # ุนููุงุช ูุฑูู ุณุงุฒ ุงูู ุชุตุงุฏู
        augmented_images = tf.image.random_flip_left_right(augmented_images)

        # ุนููุงุช ูุฑูู ุณุงุฒ ุนููุฏ ุชุตุงุฏู
        augmented_images = tf.image.random_flip_up_down(augmented_images)

        augmented_images = self.rotation_layer(augmented_images)

        return augmented_images

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

print(f"ุดฺฉู ุชุตุงูุฑ ุงุตู: {train_images.shape}")
print("\n" + "="*50 + "\n")

# ุงุฌุงุฏ ฺฉ ููููู ุงุฒ ูุงู ุงูุฒุงุด ุฏุงุฏู ู ุงุนูุงู ุขู
num_samples_to_display = 5
sample_images = train_images[:num_samples_to_display]

# ุงุฌุงุฏ ฺฉ ููููู ุงุฒ ูุงู ูุง
data_augmenter = RandomRotateFlip(rotation_factor=0.1)

# ุงุนูุงู ูุงู ุงูุฒุงุด ุฏุงุฏู ุจู ุชุตุงูุฑ ููููู
augmented_sample_images = data_augmenter(sample_images, training=True)

print(f"ุดฺฉู ุชุตุงูุฑ ุชููุช ุดุฏู: {augmented_sample_images.shape}")
print("\n" + "="*50 + "\n")

# ููุงุด ุชุตุงูุฑ ุงุตู ู ุชููุช ุดุฏู
plt.figure(figsize=(10, 5))

for i in range(num_samples_to_display):
    # ููุงุด ุชุตูุฑ ุงุตู
    plt.subplot(2, num_samples_to_display, i + 1)
    plt.imshow(sample_images[i].squeeze(), cmap='gray')
    plt.title(f"Original\nLabel: {train_labels[i]}")
    plt.axis('off')

    # ููุงุด ุชุตูุฑ ุชููุช ุดุฏู
    plt.subplot(2, num_samples_to_display, i + num_samples_to_display + 1)
    plt.imshow(augmented_sample_images[i].squeeze(), cmap='gray')
    plt.title(f"Augmented")
    plt.axis('off')

plt.tight_layout()
plt.suptitle("Original vs. Augmented Images (Rotation & Flip)", fontsize=16)
plt.show()

print("\n--- ูฺฉุชู ููู ---")
print("ุงู ูุงู ุฑู ุจู ุนููุงู ุงููู ูุงู ุฏุฑ ูุฏู Keras ุฎูุฏุชูู ูุฑุงุฑ ุจุฏุฏ:")
print("model = keras.Sequential([")
print("  RandomRotateFlip(rotation_factor=0.1),")
print("  layers.Conv2D(32, 3, activation='relu'),")
print("  # ... ุณุงุฑ ูุงูโูุง ูุฏู ุดูุง ...")
print("])")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# --- ุชุนุฑู ูุงู ุงูุฒุงุด ุฏุงุฏู ุณูุงุฑุด (ุฏูุจุงุฑู ุจุฑุง ฺฉุงูู ุจูุฏู ฺฉุฏ) ---
class RandomRotateFlip(layers.Layer):
    def __init__(self, rotation_factor=0.1, **kwargs):
        super().__init__(**kwargs)
        self.rotation_factor = rotation_factor
        self.rotation_layer = layers.RandomRotation(
            factor=rotation_factor,
            seed=None
        )
        print(f"ูุงู RandomRotateFlip ุจุง ูุงฺฉุชูุฑ ฺุฑุฎุด {rotation_factor} ุงุฌุงุฏ ุดุฏ.")

    def call(self, images, training=True):
        if not training:
            return images

        augmented_images = images
        # ูุฑููโุณุงุฒ ุงูู ุชุตุงุฏู
        augmented_images = tf.image.random_flip_left_right(augmented_images)
        # ูุฑููโุณุงุฒ ุนููุฏ ุชุตุงุฏู
        augmented_images = tf.image.random_flip_up_down(augmented_images)
        # ฺุฑุฎุด ุชุตุงุฏู
        augmented_images = self.rotation_layer(augmented_images)
        return augmented_images

# --- ุจุงุฑฺฏุฐุงุฑ ุฏุชุงุณุช MNIST (ุงู ุจุฎุด ูู ุชฺฉุฑุงุฑ ุงุณุช) ---
print("ุจุงุฑฺฏุฐุงุฑ ุฏุชุงุณุช MNIST...")
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# ูุฑูุงูโุณุงุฒ ุชุตุงูุฑ
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

# ุงุถุงูู ฺฉุฑุฏู ุจุนุฏ ฺฉุงูุงู
train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

print(f"ุดฺฉู ุชุตุงูุฑ ุขููุฒุด: {train_images.shape}")

# ุงูุชุฎุงุจ ฺูุฏ ุชุตูุฑ ููููู (batch)
num_samples_in_batch = 5 # ุชุนุฏุงุฏ ุชุตุงูุฑ ฺฉู ุฏุฑ ุงู batch ูโุฎูุงูู ููุงุด ุฏูู
print(f"\nุงูุชุฎุงุจ {num_samples_in_batch} ุชุตูุฑ ููููู ุจุฑุง ูพุฑุฏุงุฒุด ู ููุงุด...")
# ุงูุชุฎุงุจ ุชุตุงูุฑ ุงุฒ ุงุจุชุฏุง ุฏุชุงุณุช ุขููุฒุด
sample_batch_images = train_images[:num_samples_in_batch]
sample_batch_labels = train_labels[:num_samples_in_batch]

print(f"ุดฺฉู batch ุชุตุงูุฑ ููููู: {sample_batch_images.shape}")

# ุงุฌุงุฏ ูุงู ุงูุฒุงุด ุฏุงุฏู ู ุงุนูุงู ุขู ุจุฑ ุฑู batch
print("\nุงุฌุงุฏ ูุงู ุงูุฒุงุด ุฏุงุฏู ู ุงุนูุงู ุขู ุจุฑ ุฑู batch ุชุตุงูุฑ...")
# ุจุง ูุงฺฉุชูุฑ ฺุฑุฎุด 0.3 ุจุฑุง ูุงุถุญโุชุฑ ุดุฏู ุชุบุฑุงุช
augmentation_layer = RandomRotateFlip(rotation_factor=0.3)

# ุงุนูุงู ูุงู ุงูุฒุงุด ุฏุงุฏู ุฑู batch.
augmented_batch_images = augmentation_layer(sample_batch_images, training=True)

print("ุงูุฒุงุด ุฏุงุฏู ุฑู batch ุงุนูุงู ุดุฏ.")

# ููุงุด ุชุตุงูุฑ ุงุตู ู ุชููุช ุดุฏู ุฏุฑ ูุงูุจ ุดุจฺฉู
print("\nููุงุด ุชุตุงูุฑ ุงุตู ู ุชููุช ุดุฏู...")
plt.figure(figsize=(12, 8)) # ุงูุฏุงุฒู ฺฉู ุจุฑุง ููุงุด

# ุชุนุฏุงุฏ ุฑุฏูโูุง ู ุณุชููโูุง ุดุจฺฉู ููุงุด
rows = 2
cols = num_samples_in_batch

for i in range(num_samples_in_batch):
    # ููุงุด ุชุตูุฑ ุงุตู ุฏุฑ ุฑุฏู ุงูู
    plt.subplot(rows, cols, i + 1)
    # ุงุณุชูุงุฏู ุงุฒ .squeeze() ฺูู sample_batch_images ุงูุงู ฺฉ ุขุฑุงู NumPy ุงุณุช
    plt.imshow(sample_batch_images[i].squeeze(), cmap='gray')
    plt.title(f"Original\nLabel: {sample_batch_labels[i]}")
    plt.axis('off') # ุญุฐู ูุญูุฑูุง ุจุฑุง ุธุงูุฑ ุชูุฒุชุฑ

    # ููุงุด ุชุตูุฑ ุชููุช ุดุฏู ุฏุฑ ุฑุฏู ุฏูู
    plt.subplot(rows, cols, i + cols + 1)
    # ุงุณุชูุงุฏู ุงุฒ .squeeze() ฺูู augmented_batch_images ูู ฺฉ ุขุฑุงู NumPy ุงุณุช
    plt.imshow(augmented_batch_images[i].squeeze(), cmap='gray')
    plt.title(f"Augmented")
    plt.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # ุชูุธู ฺุฏูุงู ุจุฑุง ุฌููฺฏุฑ ุงุฒ ูููพูุดุงู
plt.suptitle("Comparison: Original Batch vs. Augmented Batch", fontsize=16, weight='bold')
plt.show()

print("\nููุงูุทูุฑ ฺฉู ูโุจูุฏุ ูุฑ ุชุตูุฑ ุฏุฑ batch ุจุง ุชุฑฺฉุจ ุชุตุงุฏู ฺุฑุฎุด ู ูุฑููโุณุงุฒ ุชุบุฑ ุงูุชู ุงุณุช.")
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

print("ุจุงุฑฺฏุฐุงุฑ ุฏุชุงุณุช MNIST...")
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

print(f"ููุน ุฏุงุฏู ุงุตู ุชุตุงูุฑ ุขููุฒุด: {train_images.dtype}")
print(f"ุจุงุฒู ููุงุฏุฑ ูพฺฉุณูโูุง ูุจู ุงุฒ ูุฑูุงูโุณุงุฒ: [{train_images.min()}, {train_images.max()}]")

def normalize_image(image):
    image = tf.cast(image, tf.float32)

    # ุชูุณู ุชูุงู ูพฺฉุณูโูุง ุจุฑ 255.0 ุจุฑุง ูุฑูุงูโุณุงุฒ
    normalized_image = image / 255.0

    return normalized_image

print("\nุชุงุจุน normalize_image ุจุฑุง ูุฑูุงูโุณุงุฒ ูพฺฉุณูโูุง ุชุนุฑู ุดุฏ.")

print("ุชุจุฏู ุขุฑุงูโูุง NumPy ุจู TensorFlow Dataset...")
train_dataset = tf.data.Dataset.from_tensor_slices(train_images)
test_dataset = tf.data.Dataset.from_tensor_slices(test_images)

print("ุงุนูุงู ุชุงุจุน ูุฑูุงูโุณุงุฒ ุฑู ุฏุชุงุณุช ุขููุฒุด...")

train_dataset_normalized = train_dataset.map(normalize_image, num_parallel_calls=tf.data.AUTOTUNE)

print("ุงุนูุงู ุชุงุจุน ูุฑูุงูโุณุงุฒ ุฑู ุฏุชุงุณุช ุงุนุชุจุงุฑุณูุฌ...")
test_dataset_normalized = test_dataset.map(normalize_image, num_parallel_calls=tf.data.AUTOTUNE)

print("ูุฑูุงูโุณุงุฒ ุจุง ููููุช ุงูุฌุงู ุดุฏ.")

# ุฏุฑุงูุช ฺฉ ุชุตูุฑ ูุฑูุงูโุณุงุฒ ุดุฏู ุงุฒ ุฏุชุงุณุช ุขููุฒุด
sample_normalized_image = next(iter(train_dataset_normalized.take(1)))

print(f"\nููุน ุฏุงุฏู ฺฉ ุชุตูุฑ ูุฑูุงูโุณุงุฒ ุดุฏู: {sample_normalized_image.dtype}")
print(f"ุจุงุฒู ููุงุฏุฑ ูพฺฉุณูโูุง ูพุณ ุงุฒ ูุฑูุงูโุณุงุฒ: [{tf.reduce_min(sample_normalized_image).numpy()}, {tf.reduce_max(sample_normalized_image).numpy()}]")

plt.figure(figsize=(6, 3))
plt.imshow(sample_normalized_image.numpy().squeeze(), cmap='gray')
plt.title("Sample Normalized Image")
plt.axis('off')
plt.show()

print("\nููุงูุทูุฑ ฺฉู ูุดุงูุฏู ูโฺฉูุฏุ ููุงุฏุฑ ูพฺฉุณูโูุง ุงฺฉููู ุจู 0.0 ู 1.0 ูุฑุงุฑ ุฏุงุฑูุฏ.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt

print("ุจุงุฑฺฏุฐุงุฑ ู ุขูุงุฏูโุณุงุฒ ุฏุชุงุณุช (MNIST ุจู ุนููุงู ูุซุงู)...")
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# ูุฑูุงูโุณุงุฒ ุชุตุงูุฑ
train_images = tf.cast(train_images, tf.float32) / 255.0
test_images = tf.cast(test_images, tf.float32) / 255.0

train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

print(f"ุดฺฉู ุชุตุงูุฑ ุขููุฒุด ูพุณ ุงุฒ ุขูุงุฏูโุณุงุฒ: {train_images.shape}") # ุจุงุฏ (60000, 28, 28, 1) ุจุงุดุฏ

#  ุชุนุฑู ูพุงุฑุงูุชุฑูุง ูุฏู
input_shape = train_images.shape[1:] # (28, 28, 1) ุจุฑุง MNIST
num_classes = 10

print("ุฏุฑ ุญุงู ุณุงุฎุช ูุฏู CNN...")
model = Sequential([
    # ูุงู ุงูู ฺฉุงููููุดู ู ูพูููฺฏ
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    MaxPooling2D((2, 2)), # ฺฉุงูุด ุงุจุนุงุฏ ุจู ูุตู

    # ูุงู ุฏูู ฺฉุงููููุดู ู ูพูููฺฏ
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # ูุงู ุณูู ฺฉุงููููุดู
    Conv2D(128, (3, 3), activation='relu'),

    # ุตุงู ฺฉุฑุฏู ุฎุฑูุฌ ูุงูโูุง ฺฉุงููููุดู ุจุฑุง ูุฑูุฏ ุจู ูุงูโูุง ุชูุงูโูุชุตู
    Flatten(),

    # ูุงู ุชูุงูโูุชุตู (Fully Connected)
    Dense(128, activation='relu'),


    Dropout(0.5), # ูุฑุฎ Dropout ุฑุง ูโุชูุงู ุชูุธู ฺฉุฑุฏ

    Dense(num_classes, activation='softmax')
])

print("ูุฏู CNN ุจุง ููููุช ุณุงุฎุชู ุดุฏ.")
model.summary() # ููุงุด ูุนูุงุฑ ูุฏู

print("\nุฏุฑ ุญุงู ฺฉุงููพุงู ฺฉุฑุฏู ูุฏู...")

# ุงุณุชูุงุฏู ุงุฒ Adam optimizer ุจุง ูุฑุฎ ุงุฏฺฏุฑ 0.001
optimizer = Adam(learning_rate=0.001)

model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("ูุฏู ุจุง ููููุช ฺฉุงููพุงู ุดุฏ ู ุขูุงุฏู ุขููุฒุด ุงุณุช.")

# ุขูุงุฏูโุณุงุฒ ุฏุงุฏูโูุง ุจุฑุง ุขููุฒุด (ุงุถุงูู ฺฉุฑุฏู batch ู prefetch)
batch_size = 64
buffer_size = tf.data.AUTOTUNE # ุงุณุชูุงุฏู ุงุฒ ุชูุธูุงุช ุฎูุฏฺฉุงุฑ TensorFlow

# ุชุจุฏู ุฏุงุฏูโูุง ูุฑูุงูโุณุงุฒ ุดุฏู ุจู tf.data.Dataset
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, test_labels))

# ุงุนูุงู batching, shuffling ู prefetching
train_dataset_processed = train_dataset.shuffle(buffer_size=buffer_size).batch(batch_size).prefetch(buffer_size=buffer_size)
test_dataset_processed = test_dataset.batch(batch_size).prefetch(buffer_size=buffer_size)

print(f"\nุฏุชุงุณุชโูุง ุจุฑุง ุขููุฒุด ุจุง batch size {batch_size} ุขูุงุฏู ุดุฏูุฏ.")
```

```python
# TODO: ุณุงุฎุช ูุฏู ุดุจฺฉู ุนุตุจ ฺฉุงููููุดู (CNN) ุจุฑุง ุทุจููโุจูุฏ ุชุตุงูุฑ

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt

# ุขูุงุฏูโุณุงุฒ ู ูพุดโูพุฑุฏุงุฒุด ุฏุงุฏูโูุง
print("ุดุฑูุน ุจุงุฑฺฏุฐุงุฑ ู ุขูุงุฏูโุณุงุฒ ุฏุงุฏูโูุง...")

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# ูุฑูุงูโุณุงุฒ ุชุตุงูุฑ:
train_images = tf.cast(train_images, tf.float32) / 255.0
test_images = tf.cast(test_images, tf.float32) / 255.0

train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

print(f"ุดฺฉู ุชุตุงูุฑ ุขููุฒุด ูพุณ ุงุฒ ุขูุงุฏูโุณุงุฒ: {train_images.shape}")
print(f"ุดฺฉู ุจุฑฺุณุจโูุง ุขููุฒุด: {train_labels.shape}")

print("\nุดุฑูุน ุทุฑุงุญ ูุนูุงุฑ ูุฏู CNN...")

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),

    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'), # 64 ููุชุฑ ุจุฑุง ุงุณุชุฎุฑุงุฌ ูฺฺฏโูุง ูพฺุฏูโุชุฑ
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    Flatten(),

    Dense(128, activation='relu'),

    Dropout(0.5),

    Dense(10, activation='softmax')
], name="Simple_CNN_Classifier") # ูุงู ูุฏู

print("ูุนูุงุฑ ูุฏู ุจุง ููููุช ุณุงุฎุชู ุดุฏ.")

model.summary()

print("\nุดุฑูุน ฺฉุงููพุงู ฺฉุฑุฏู ูุฏู...")

optimizer = Adam(learning_rate=0.001) # ูุฑุฎ ุงุฏฺฏุฑ 0.001

loss_function = 'sparse_categorical_crossentropy'

metrics_to_track = ['accuracy']

model.compile(optimizer=optimizer,
              loss=loss_function,
              metrics=metrics_to_track)

print("ูุฏู ุจุง ููููุช ฺฉุงููพุงู ุดุฏ ู ุขูุงุฏู ุขููุฒุด ุงุณุช.")

print("\nุดุฑูุน ุขููุฒุด ูุฏู...")

batch_size = 64
epochs_to_train = 10

history = model.fit(train_images, train_labels,
                    batch_size=batch_size,
                    epochs=epochs_to_train,
                    validation_data=(test_images, test_labels))

print("ุขููุฒุด ูุฏู ุจู ูพุงุงู ุฑุณุฏ.")

print("\nุงุฑุฒุงุจ ูุฏู ุจุง ุฏุงุฏูโูุง ุชุณุช...")
loss, accuracy = model.evaluate(test_images, test_labels, verbose=0)

print(f"\nูุชุฌู ุงุฑุฒุงุจ ููุง:")
print(f"  - Loss (ุฒุงู): {loss:.4f}")
print(f"  - Accuracy (ุฏูุช): {accuracy:.4f}")

print("\nููุงุด ูููุฏุงุฑูุง ุขููุฒุด...")
plt.figure(figsize=(12, 4))

# ูููุฏุงุฑ ุฏูุช (Accuracy)
plt.subplot(1, 2, 1) # 1 ุฑุฏูุ 2 ุณุชููุ ูููุฏุงุฑ ุงูู
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# ูููุฏุงุฑ ุฒุงู (Loss)
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout() # ุชูุธู ูุงุตูู ุจู ูููุฏุงุฑูุง
plt.show()

print("\nูุฑุขูุฏ ฺฉุงูู ุดุฏ! ุดูุง ฺฉ ูุฏู CNN ุณุงุฎุชุฏ ู ุขููุฒุด ุฏุงุฏุฏ.")
```

```python
# TODO: ูุฏู ุฑุง ุขููุฒุด ุฏูุฏ
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
print("ุฏุฑ ุญุงู ุขูุงุฏูโุณุงุฒ ุฏุงุฏูโูุง ุดุจูโุณุงุฒ ุดุฏู...")

num_train_samples = 1000
num_val_samples = 200

img_rows, img_cols = 28, 28
num_channels = 1
num_classes = 10

x_train = np.random.rand(num_train_samples, img_rows, img_cols, num_channels).astype('float32')
y_train = np.random.randint(0, num_classes, num_train_samples)

x_val = np.random.rand(num_val_samples, img_rows, img_cols, num_channels).astype('float32')
y_val = np.random.randint(0, num_classes, num_val_samples)

print(f"ุดฺฉู ุฏุงุฏูโูุง ุขููุฒุด (x_train): {x_train.shape}")
print(f"ุดฺฉู ุจุฑฺุณุจโูุง ุขููุฒุด (y_train): {y_train.shape}")
print(f"ุดฺฉู ุฏุงุฏูโูุง ุงุนุชุจุงุฑุณูุฌ (x_val): {x_val.shape}")
print(f"ุดฺฉู ุจุฑฺุณุจโูุง ุงุนุชุจุงุฑุณูุฌ (y_val): {y_val.shape}")

print("\nุณุงุฎุช ู ฺฉุงููพุงู ฺฉุฑุฏู ูุฏู...")

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_rows, img_cols, num_channels)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("ูุฏู ุณุงุฎุชู ู ฺฉุงููพุงู ุดุฏ.")
model.summary() # ููุงุด ุฎูุงุตู ูุฏู

print("\nุดุฑูุน ูุฑุขูุฏ ุขููุฒุด ูุฏู...")

history = model.fit(
    x=x_train,
    y=y_train,
    epochs=5,
    batch_size=32,
    validation_data=(x_val, y_val)
)

print("ูุฑุขูุฏ ุขููุฒุด ูุฏู ุจู ูพุงุงู ุฑุณุฏ!")

print("\nูุชุงุฌ ุขููุฒุด:")
print(history.history.keys())

print(f"ุขุฎุฑู ุฏูุช ุขููุฒุด: {history.history['accuracy'][-1]:.4f}")
print(f"ุขุฎุฑู ุฏูุช ุงุนุชุจุงุฑุณูุฌ: {history.history['val_accuracy'][-1]:.4f}")

```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt # ูุงุฑุฏ ฺฉุฑุฏู ฺฉุชุงุจุฎุงูู ุฑุณู ูููุฏุงุฑ

print("ุฏุฑ ุญุงู ุขูุงุฏูโุณุงุฒ ุฏุงุฏูโูุง ุดุจูโุณุงุฒ ุดุฏู...")

num_train_samples = 1000
num_val_samples = 200
img_rows, img_cols = 28, 28
num_channels = 1
num_classes = 10

x_train = np.random.rand(num_train_samples, img_rows, img_cols, num_channels).astype('float32')
y_train = np.random.randint(0, num_classes, num_train_samples)

x_val = np.random.rand(num_val_samples, img_rows, img_cols, num_channels).astype('float32')
y_val = np.random.randint(0, num_classes, num_val_samples)

print(f"ุดฺฉู ุฏุงุฏูโูุง ุขููุฒุด (x_train): {x_train.shape}")
print(f"ุดฺฉู ุจุฑฺุณุจโูุง ุขููุฒุด (y_train): {y_train.shape}")
print(f"ุดฺฉู ุฏุงุฏูโูุง ุงุนุชุจุงุฑุณูุฌ (x_val): {x_val.shape}")
print(f"ุดฺฉู ุจุฑฺุณุจโูุง ุงุนุชุจุงุฑุณูุฌ (y_val): {y_val.shape}")

print("\nุณุงุฎุช ู ฺฉุงููพุงู ฺฉุฑุฏู ูุฏู...")

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_rows, img_cols, num_channels)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("ูุฏู ุณุงุฎุชู ู ฺฉุงููพุงู ุดุฏ.")

print("\nุดุฑูุน ูุฑุขูุฏ ุขููุฒุด ูุฏู...")

history = model.fit(
    x=x_train,
    y=y_train,
    epochs=10,
    batch_size=32,
    validation_data=(x_val, y_val)
)

print("ูุฑุขูุฏ ุขููุฒุด ูุฏู ุจู ูพุงุงู ุฑุณุฏ!")

print("\nุฏุฑ ุญุงู ุฑุณู ูููุฏุงุฑูุง Loss ู Accuracy...")

# ุฏุณุชุฑุณ ุจู ููุงุฏุฑ Loss ู Accuracy ุงุฒ ุดุก history
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(len(acc))

plt.figure(figsize=(12, 8)) # ุงูุฏุงุฒู ฺฉู ุดฺฉู ูููุฏุงุฑ

plt.subplot(1, 2, 1)
plt.plot(epochs_range, loss, label='Loss (ุขููุฒุด)')
plt.plot(epochs_range, val_loss, label='Loss (ุงุนุชุจุงุฑุณูุฌ)')
plt.legend(loc='upper right')
plt.title('ูููุฏุงุฑ Loss')
plt.xlabel('Epoch')
plt.ylabel('ููุฏุงุฑ Loss')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, acc, label='Accuracy (ุขููุฒุด)')
plt.plot(epochs_range, val_acc, label='Accuracy (ุงุนุชุจุงุฑุณูุฌ)')
plt.legend(loc='lower right')
plt.title('ูููุฏุงุฑ Accuracy')
plt.xlabel('Epoch')
plt.ylabel('ููุฏุงุฑ Accuracy')

# ููุงุด ูููุฏุงุฑูุง
plt.tight_layout()
plt.show()

print("ูููุฏุงุฑูุง ุจุง ููููุช ููุงุด ุฏุงุฏู ุดุฏูุฏ.")

# ููุงุด ูุชุงุฌ ููุง (ุงุฎุชุงุฑ)
print("\nูุชุงุฌ ููุง ุขููุฒุด:")
print(f"ุขุฎุฑู ุฏูุช ุขููุฒุด: {acc[-1]:.4f}")
print(f"ุขุฎุฑู ุฏูุช ุงุนุชุจุงุฑุณูุฌ: {val_acc[-1]:.4f}")
print(f"ุขุฎุฑู Loss ุขููุฒุด: {loss[-1]:.4f}")
print(f"ุขุฎุฑู Loss ุงุนุชุจุงุฑุณูุฌ: {val_loss[-1]:.4f}")

```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

print("ุณุงุฎุช ู ฺฉุงููพุงู ฺฉุฑุฏู ูุฏู (ูุดุงุจู ูุจู)...")
num_train_samples = 1000
num_val_samples = 200
img_rows, img_cols = 28, 28
num_channels = 1
num_classes = 10

x_train = np.random.rand(num_train_samples, img_rows, img_cols, num_channels).astype('float32')
y_train = np.random.randint(0, num_classes, num_train_samples)
x_val = np.random.rand(num_val_samples, img_rows, img_cols, num_channels).astype('float32')
y_val = np.random.randint(0, num_classes, num_val_samples)

# ูุฏู CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_rows, img_cols, num_channels)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("ุขููุฒุด ูุฏู ุจุง ุฏุงุฏูโูุง ุดุจูโุณุงุฒ ุดุฏู...")
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=5,
    batch_size=32,
    validation_data=(x_val, y_val),
    verbose=0 # ุฎุงููุด ฺฉุฑุฏู ููุงุด ุฌุฒุฆุงุช ุขููุฒุด
)
print("ูุฏู ุจุง ููููุช ุขููุฒุด ุฏุฏ.")

print("\nุขูุงุฏูโุณุงุฒ ฺฉ ุชุตูุฑ ุฌุฏุฏ ุจุฑุง ูพุดโุจู...")

new_image_data = np.random.rand(img_rows, img_cols, num_channels).astype('float32')

input_for_prediction = np.expand_dims(new_image_data, axis=0)

print(f"ุดฺฉู ุชุตูุฑ ูุฑูุฏ ุจุฑุง ูพุดโุจู: {input_for_prediction.shape}")

print("\nุงูุฌุงู ูพุดโุจู ฺฉูุงุณ ุจุฑุง ุชุตูุฑ ุฌุฏุฏ...")

predictions = model.predict(input_for_prediction)

print(f"ุฎุฑูุฌ ูพุดโุจู (ุงุญุชูุงูุงุช ฺฉูุงุณโูุง): {predictions}")

# ูพุฏุง ฺฉุฑุฏู ุดุงุฎุต ฺฉูุงุณ ฺฉู ุจุงูุงุชุฑู ุงุญุชูุงู ุฑุง ุฏุงุฑุฏ
predicted_class_index = np.argmax(predictions[0])

# ฺฏุฑูุชู ุจุงูุงุชุฑู ุงุญุชูุงู
confidence = np.max(predictions[0])

print(f"\nฺฉูุงุณ ูพุดโุจู ุดุฏู: {predicted_class_index}")
print(f"ูุฒุงู ุงุทููุงู ูุฏู: {confidence:.4f}")

print("\nููุงุด ุชุตูุฑ ูุฑูุฏ ู ูุชุฌู ูพุดโุจู...")

plt.figure(figsize=(6, 6))
plt.imshow(new_image_data.squeeze(), cmap='gray')
plt.title(f"ูพุดโุจู ุดุฏู: {predicted_class_index}\nุงุทููุงู: {confidence:.4f}")
plt.axis('off') # ุญุฐู ูุญูุฑูุง
plt.show()

print("ูพุดโุจู ู ููุงุด ุชุตูุฑ ุงูุฌุงู ุดุฏ.")
```
