```python
# from tensorflow.keras.preprocessing.image import ImageDataGenerator

# پارامترها
IMG_HEIGHT = 150
IMG_WIDTH = 150
BATCH_SIZE = 32
VALIDATION_SPLIT = 0.2  

# ایجاد ImageDataGenerator برای augmentation و normalization
datagen = ImageDataGenerator(
    rescale=1./255,  
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=VALIDATION_SPLIT  
)

# تولید جریان داده آموزشی
train_generator = datagen.flow_from_directory(
    BASE_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training' 
)

# تولید جریان داده validation
validation_generator = datagen.flow_from_directory(
    BASE_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'  
)

# بررسی کلاس‌ها
print("کلاس‌ها:", train_generator.class_indices)
```

```python
# ایجاد مدل
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(len(class_names), activation='softmax')  
])

# کامپایل مدل
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# نمایش خلاصه مدل
model.summary()

# آموزش مدل
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE
)
```

```python
# TODO: اندازه بچ و لیبل ها را در خروجی نشان دهید.

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# شبیه‌سازی دیتاست 
num_samples = 10
input_dimension = 100 
num_classes = 3 

X_data = np.random.rand(num_samples, input_dimension).astype(np.float32)

y_labels_true = np.random.randint(0, num_classes, size=num_samples)

print(f"شکل داده‌های ورودی (X_data): {X_data.shape}")
print(f"شکل لیبل‌های واقعی (y_labels_true): {y_labels_true.shape}")
print(f"نمونه‌ای از داده‌های ورودی (X_data[0]):\n{X_data[0][:5]}...") 
print(f"نمونه‌ای از لیبل‌های واقعی (y_labels_true):\n{y_labels_true}")

print("\n" + "="*50 + "\n")
```

```python
# تعریف یک مدل عصبی ساده 
model = keras.Sequential([
    layers.Dense(units=64, activation='relu', input_shape=(input_dimension,)),
    layers.Dense(units=32, activation='relu'),
    layers.Dense(units=num_classes, activation='softmax') 
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

print("مدل عصبی ساده تعریف شد:")
model.summary()
print("\n" + "="*50 + "\n")

#  آموزش مدل با استفاده از بچ 
print("شروع آموزش مدل با استفاده از Batch Size...")

batch_size = 4 
epochs = 2     

history = model.fit(X_data, y_labels_true,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1) 

print(f"\nآموزش به پایان رسید. (هر Epoch شامل {np.ceil(num_samples / batch_size).astype(int)} گام/بچ بود.)")
print("در حین آموزش، مدل داده‌ها را به بچ‌های 4 تایی دریافت و پردازش کرد.")
print("\n" + "="*50 + "\n")
```

```python
#  پیش‌بینی با مدل 
print("شروع پیش‌بینی با مدل...")

num_predictions = 3
X_to_predict = X_data[:num_predictions]
y_true_for_prediction = y_labels_true[:num_predictions]

predictions = model.predict(X_to_predict)

print(f"\nشکل خروجی پیش‌بینی‌ها: {predictions.shape}") 
print("\nخروجی احتمالات (Score) برای هر کلاس (نمونه‌های اصلی X_data):")
print(predictions)

print("\n نحوه تفسیر خروجی ")
for i in range(num_predictions):
    print(f"\nنمونه {i+1}:")
    print(f"  لیبل واقعی: {y_true_for_prediction[i]}")
    print(f"  احتمالات پیش‌بینی شده توسط مدل: {predictions[i]}")

    # برای به دست آوردن لیبل پیش‌بینی شده، از argmax استفاده می‌کنیم
    predicted_label = np.argmax(predictions[i])
    print(f"  لیبل پیش‌بینی شده توسط مدل (با بالاترین احتمال): {predicted_label}")

    # نمایش نام کلاس‌ها برای فهم بهتر
    class_map = {0: "سنگ", 1: "کاغذ", 2: "قیچی"}
    print(f"  نام لیبل واقعی: {class_map[y_true_for_prediction[i]]}")
    print(f"  نام لیبل پیش‌بینی شده: {class_map[predicted_label]}")

print("\n" + "="*50 + "\n")

print("مثال: پیش‌بینی برای کل دیتاست (باز هم به صورت بچ‌های داخلی انجام می‌شود):")
all_predictions = model.predict(X_data)
print(f"شکل خروجی پیش‌بینی برای کل دیتاست: {all_predictions.shape}")
print(f"نمونه خروجی برای اولین نمونه (احتمالات): {all_predictions[0]}")
print(f"لیبل پیش‌بینی شده برای اولین نمونه: {np.argmax(all_predictions[0])}")
```

        <li>        از .shuffle() برای ایجاد تنوع در ترتیب داده‌های آموزشی استفاده کنید تا مدل دچار overfitting نشود.</li>
        <li>        از .prefetch() برای بارگذاری داده‌ها به‌صورت غیرهمزمان استفاده کنید تا بهره‌وری پردازنده/کارت گرافیک افزایش یابد.</li>

```python
# TODO : کش کردن داده‌ها در حافظه برای افزایش سرعت آموزش (فقط در صورت کافی بودن حافظه RAM)
import tensorflow as tf
import numpy as np
import time

num_samples = 10000
input_dim = 100 # ابعاد هر نمونه 
num_classes = 10 # تعداد کلاس‌ها

X_data_simulated = np.random.rand(num_samples, input_dim).astype(np.float32)
y_labels_simulated = np.random.randint(0, num_classes, size=num_samples)

print(f"شکل داده‌های شبیه‌سازی شده (X_data_simulated): {X_data_simulated.shape}")
print(f"شکل لیبل‌های شبیه‌سازی شده (y_labels_simulated): {y_labels_simulated.shape}")
print("\n" + "="*50 + "\n")

def load_data_slowly(feature, label):

    time.sleep(0.0001) # شبیه‌سازی تاخیر 0.1 میلی‌ثانیه‌ای برای هر نمونه
    return feature, label

# ایجاد tf.data.Dataset از داده‌های NumPy
dataset_original = tf.data.Dataset.from_tensor_slices((X_data_simulated, y_labels_simulated))

# نگاشت تابع load_data_slowly به هر نمونه
dataset_processed = dataset_original.map(load_data_slowly)

# شافل کردن و بچ کردن
batch_size = 32
dataset_batched = dataset_processed.shuffle(buffer_size=1000).batch(batch_size)

print("پایپ‌لاین داده بدون کش ایجاد شد.")
print("\n" + "="*50 + "\n")

# 3. اندازه‌گیری زمان آموزش (بدون کش) 
print("شروع آموزش بدون کش...")
start_time_no_cache = time.time()

epochs = 3 

# شبیه‌سازی حلقه آموزش
for epoch in range(epochs):
    print(f"Epoch {epoch+1} (بدون کش):")
    for batch_features, batch_labels in dataset_batched:
            
        _ = 1  
    print(f"  Epoch {epoch+1} کامل شد.")

end_time_no_cache = time.time()
time_taken_no_cache = end_time_no_cache - start_time_no_cache
print(f"\nزمان کل آموزش بدون کش: {time_taken_no_cache:.2f} ثانیه")

print("\n" + "="*50 + "\n")

dataset_cached = dataset_original.map(load_data_slowly).cache()

# شافل کردن و بچ کردن 
dataset_cached_batched = dataset_cached.shuffle(buffer_size=1000).batch(batch_size)

print("پایپ‌لاین داده با قابلیت کش ایجاد شد.")
print("\n" + "="*50 + "\n")

print("شروع آموزش با کش...")
start_time_cache = time.time()

# شبیه‌سازی حلقه آموزش
for epoch in range(epochs):
    print(f"Epoch {epoch+1} (با کش):")
    for batch_features, batch_labels in dataset_cached_batched:
        _ = 1  
    print(f"  Epoch {epoch+1} کامل شد.")

end_time_cache = time.time()
time_taken_cache = end_time_cache - start_time_cache
print(f"\nزمان کل آموزش با کش: {time_taken_cache:.2f} ثانیه")

print("\n" + "="*50 + "\n")

#  مقایسه نتایج 
print("--- مقایسه نتایج ---")
print(f"زمان بدون کش: {time_taken_no_cache:.2f} ثانیه")
print(f"زمان با کش: {time_taken_cache:.2f} ثانیه")
if time_taken_no_cache > time_taken_cache:
    print(f"کش کردن باعث سرعت بخشیدن به آموزش شد! (حدود {time_taken_no_cache / time_taken_cache:.2f} برابر سریعتر)")
else:
    print("کش کردن در این مثال بهبود قابل توجهی ایجاد نکرد (ممکن است به دلیل دیتاست کوچک یا سرعت بالای I/O شبیه‌سازی شده باشد).")
```

      هدف این مرحله افزایش تنوع داده‌های آموزشی برای بهبود تعمیم‌پذیری مدل است.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# تعریف لایه افزایش داده سفارشی اصلاح شده 
class RandomRotateFlip(layers.Layer):
    def __init__(self, rotation_factor=0.1, **kwargs):
        super().__init__(**kwargs)
        
        
        self.rotation_factor = rotation_factor

        # ایجاد لایه RandomRotation در زمان ساخت لایه سفارشی
        self.rotation_layer = layers.RandomRotation(
            factor=rotation_factor,  
            seed=None  
        )

    def call(self, images, training=True):
        if not training:
            return images 

        augmented_images = images

        # عملیات قرینه سازی افقی تصادفی 
        augmented_images = tf.image.random_flip_left_right(augmented_images)

        # عملیات قرینه سازی عمودی تصادفی 
        augmented_images = tf.image.random_flip_up_down(augmented_images)

        augmented_images = self.rotation_layer(augmented_images)

        return augmented_images

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

print(f"شکل تصاویر اصلی: {train_images.shape}")
print("\n" + "="*50 + "\n")

# ایجاد یک نمونه از لایه افزایش داده و اعمال آن 
num_samples_to_display = 5
sample_images = train_images[:num_samples_to_display]

# ایجاد یک نمونه از لایه ما
data_augmenter = RandomRotateFlip(rotation_factor=0.1)

# اعمال لایه افزایش داده به تصاویر نمونه
augmented_sample_images = data_augmenter(sample_images, training=True)

print(f"شکل تصاویر تقویت شده: {augmented_sample_images.shape}")
print("\n" + "="*50 + "\n")

# نمایش تصاویر اصلی و تقویت شده 
plt.figure(figsize=(10, 5))

for i in range(num_samples_to_display):
    # نمایش تصویر اصلی
    plt.subplot(2, num_samples_to_display, i + 1)
    plt.imshow(sample_images[i].squeeze(), cmap='gray')
    plt.title(f"Original\nLabel: {train_labels[i]}")
    plt.axis('off')

    # نمایش تصویر تقویت شده
    plt.subplot(2, num_samples_to_display, i + num_samples_to_display + 1)
    plt.imshow(augmented_sample_images[i].squeeze(), cmap='gray')
    plt.title(f"Augmented")
    plt.axis('off')

plt.tight_layout()
plt.suptitle("Original vs. Augmented Images (Rotation & Flip)", fontsize=16)
plt.show()

print("\n--- نکته مهم ---")
print("این لایه رو به عنوان اولین لایه در مدل Keras خودتون قرار بدید:")
print("model = keras.Sequential([")
print("  RandomRotateFlip(rotation_factor=0.1),")
print("  layers.Conv2D(32, 3, activation='relu'),")
print("  # ... سایر لایه‌های مدل شما ...")
print("])")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# --- تعریف لایه افزایش داده سفارشی (دوباره برای کامل بودن کد) ---
class RandomRotateFlip(layers.Layer):
    def __init__(self, rotation_factor=0.1, **kwargs):
        super().__init__(**kwargs)
        self.rotation_factor = rotation_factor
        self.rotation_layer = layers.RandomRotation(
            factor=rotation_factor,
            seed=None
        )
        print(f"لایه RandomRotateFlip با فاکتور چرخش {rotation_factor} ایجاد شد.")

    def call(self, images, training=True):
        if not training:
            return images

        augmented_images = images
        # قرینه‌سازی افقی تصادفی
        augmented_images = tf.image.random_flip_left_right(augmented_images)
        # قرینه‌سازی عمودی تصادفی
        augmented_images = tf.image.random_flip_up_down(augmented_images)
        # چرخش تصادفی
        augmented_images = self.rotation_layer(augmented_images)
        return augmented_images

# --- بارگذاری دیتاست MNIST (این بخش هم تکراری است) ---
print("بارگذاری دیتاست MNIST...")
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# نرمال‌سازی تصاویر
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

# اضافه کردن بعد کانال
train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

print(f"شکل تصاویر آموزشی: {train_images.shape}")

# انتخاب چند تصویر نمونه (batch) 
num_samples_in_batch = 5 # تعداد تصاویری که در این batch می‌خواهم نمایش دهیم
print(f"\nانتخاب {num_samples_in_batch} تصویر نمونه برای پردازش و نمایش...")
# انتخاب تصاویر از ابتدای دیتاست آموزشی
sample_batch_images = train_images[:num_samples_in_batch]
sample_batch_labels = train_labels[:num_samples_in_batch]

print(f"شکل batch تصاویر نمونه: {sample_batch_images.shape}")

# ایجاد لایه افزایش داده و اعمال آن بر روی batch 
print("\nایجاد لایه افزایش داده و اعمال آن بر روی batch تصاویر...")
# با فاکتور چرخش 0.3 برای واضح‌تر شدن تغییرات
augmentation_layer = RandomRotateFlip(rotation_factor=0.3)

# اعمال لایه افزایش داده روی batch. 
augmented_batch_images = augmentation_layer(sample_batch_images, training=True)

print("افزایش داده روی batch اعمال شد.")

# نمایش تصاویر اصلی و تقویت شده در قالب شبکه 
print("\nنمایش تصاویر اصلی و تقویت شده...")
plt.figure(figsize=(12, 8)) # اندازه کلی برای نمایش

# تعداد ردیف‌ها و ستون‌های شبکه نمایش
rows = 2
cols = num_samples_in_batch

for i in range(num_samples_in_batch):
    # نمایش تصویر اصلی در ردیف اول
    plt.subplot(rows, cols, i + 1)
    # استفاده از .squeeze() چون sample_batch_images الان یک آرایه NumPy است
    plt.imshow(sample_batch_images[i].squeeze(), cmap='gray')
    plt.title(f"Original\nLabel: {sample_batch_labels[i]}")
    plt.axis('off') # حذف محورها برای ظاهر تمیزتر

    # نمایش تصویر تقویت شده در ردیف دوم
    plt.subplot(rows, cols, i + cols + 1)
    # استفاده از .squeeze() چون augmented_batch_images هم یک آرایه NumPy است
    plt.imshow(augmented_batch_images[i].squeeze(), cmap='gray')
    plt.title(f"Augmented")
    plt.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # تنظیم چیدمان برای جلوگیری از همپوشانی
plt.suptitle("Comparison: Original Batch vs. Augmented Batch", fontsize=16, weight='bold')
plt.show()

print("\nهمانطور که می‌بینید، هر تصویر در batch با ترکیب تصادفی چرخش و قرینه‌سازی تغییر یافته است.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt

print("بارگذاری و آماده‌سازی دیتاست (MNIST به عنوان مثال)...")
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# نرمال‌سازی تصاویر 
train_images = tf.cast(train_images, tf.float32) / 255.0
test_images = tf.cast(test_images, tf.float32) / 255.0

train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

print(f"شکل تصاویر آموزشی پس از آماده‌سازی: {train_images.shape}") # باید (60000, 28, 28, 1) باشد

#  تعریف پارامترهای مدل 
input_shape = train_images.shape[1:] # (28, 28, 1) برای MNIST
num_classes = 10 

print("در حال ساخت مدل CNN...")
model = Sequential([
    # لایه اول کانولوشن و پولینگ
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape), 
    MaxPooling2D((2, 2)), # کاهش ابعاد به نصف

    # لایه دوم کانولوشن و پولینگ
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # لایه سوم کانولوشن 
    Conv2D(128, (3, 3), activation='relu'),
    
    # صاف کردن خروجی لایه‌های کانولوشن برای ورود به لایه‌های تمام‌متصل
    Flatten(),

    # لایه تمام‌متصل (Fully Connected)
    Dense(128, activation='relu'),


    Dropout(0.5), # نرخ Dropout را می‌توان تنظیم کرد

    Dense(num_classes, activation='softmax')
])

print("مدل CNN با موفقیت ساخته شد.")
model.summary() # نمایش معماری مدل

print("\nدر حال کامپایل کردن مدل...")

# استفاده از Adam optimizer با نرخ یادگیری 0.001
optimizer = Adam(learning_rate=0.001)

model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

print("مدل با موفقیت کامپایل شد و آماده آموزش است.")

# آماده‌سازی داده‌ها برای آموزش (اضافه کردن batch و prefetch) 
batch_size = 64
buffer_size = tf.data.AUTOTUNE # استفاده از تنظیمات خودکار TensorFlow

# تبدیل داده‌های نرمال‌سازی شده به tf.data.Dataset
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, test_labels))

# اعمال batching, shuffling و prefetching
train_dataset_processed = train_dataset.shuffle(buffer_size=buffer_size).batch(batch_size).prefetch(buffer_size=buffer_size)
test_dataset_processed = test_dataset.batch(batch_size).prefetch(buffer_size=buffer_size)

print(f"\nدیتاست‌ها برای آموزش با batch size {batch_size} آماده شدند.")
```

```python
# TODO: ساخت مدل شبکه عصبی کانولوشنی (CNN) برای طبقه‌بندی تصاویر

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt

# آماده‌سازی و پیش‌پردازش داده‌ها 
print("شروع بارگذاری و آماده‌سازی داده‌ها...")

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# نرمال‌سازی تصاویر:
train_images = tf.cast(train_images, tf.float32) / 255.0
test_images = tf.cast(test_images, tf.float32) / 255.0

train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

print(f"شکل تصاویر آموزشی پس از آماده‌سازی: {train_images.shape}") 
print(f"شکل برچسب‌های آموزشی: {train_labels.shape}") 

print("\nشروع طراحی معماری مدل CNN...")

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),

    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'), # 64 فیلتر برای استخراج ویژگی‌های پیچیده‌تر
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    Flatten(),

    Dense(128, activation='relu'),

    Dropout(0.5),

    Dense(10, activation='softmax')
], name="Simple_CNN_Classifier") # نام مدل

print("معماری مدل با موفقیت ساخته شد.")

model.summary()

print("\nشروع کامپایل کردن مدل...")

optimizer = Adam(learning_rate=0.001) # نرخ یادگیری 0.001

loss_function = 'sparse_categorical_crossentropy'

metrics_to_track = ['accuracy']

model.compile(optimizer=optimizer,
              loss=loss_function,
              metrics=metrics_to_track)

print("مدل با موفقیت کامپایل شد و آماده آموزش است.")

print("\nشروع آموزش مدل...")

batch_size = 64
epochs_to_train = 10 

history = model.fit(train_images, train_labels,
                    batch_size=batch_size,
                    epochs=epochs_to_train,
                    validation_data=(test_images, test_labels))

print("آموزش مدل به پایان رسید.")

print("\nارزیابی مدل با داده‌های تست...")
loss, accuracy = model.evaluate(test_images, test_labels, verbose=0)

print(f"\nنتیجه ارزیابی نهایی:")
print(f"  - Loss (زیان): {loss:.4f}")
print(f"  - Accuracy (دقت): {accuracy:.4f}")

print("\nنمایش نمودارهای آموزش...")
plt.figure(figsize=(12, 4))

# نمودار دقت (Accuracy)
plt.subplot(1, 2, 1) # 1 ردیف، 2 ستون، نمودار اول
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# نمودار زیان (Loss)
plt.subplot(1, 2, 2) 
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout() # تنظیم فاصله بین نمودارها
plt.show()

print("\nفرآیند کامل شد! شما یک مدل CNN ساختید و آموزش دادید.")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

print("ساخت و کامپایل کردن مدل (مشابه قبل)...")
num_train_samples = 1000
num_val_samples = 200
img_rows, img_cols = 28, 28
num_channels = 1
num_classes = 10

x_train = np.random.rand(num_train_samples, img_rows, img_cols, num_channels).astype('float32')
y_train = np.random.randint(0, num_classes, num_train_samples)
x_val = np.random.rand(num_val_samples, img_rows, img_cols, num_channels).astype('float32')
y_val = np.random.randint(0, num_classes, num_val_samples)

# مدل CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_rows, img_cols, num_channels)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("آموزش مدل با داده‌های شبیه‌سازی شده...")
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=5, 
    batch_size=32,
    validation_data=(x_val, y_val),
    verbose=0 # خاموش کردن نمایش جزئیات آموزش 
)
print("مدل با موفقیت آموزش دید.")

print("\nآماده‌سازی یک تصویر جدید برای پیش‌بینی...")

new_image_data = np.random.rand(img_rows, img_cols, num_channels).astype('float32')

input_for_prediction = np.expand_dims(new_image_data, axis=0)

print(f"شکل تصویر ورودی برای پیش‌بینی: {input_for_prediction.shape}")

print("\nانجام پیش‌بینی کلاس برای تصویر جدید...")

predictions = model.predict(input_for_prediction)

print(f"خروجی پیش‌بینی (احتمالات کلاس‌ها): {predictions}")

# پیدا کردن شاخص کلاسی که بالاترین احتمال را دارد
predicted_class_index = np.argmax(predictions[0])

# گرفتن بالاترین احتمال 
confidence = np.max(predictions[0])

print(f"\nکلاس پیش‌بینی شده: {predicted_class_index}")
print(f"میزان اطمینان مدل: {confidence:.4f}")

print("\nنمایش تصویر ورودی و نتیجه پیش‌بینی...")

plt.figure(figsize=(6, 6))
plt.imshow(new_image_data.squeeze(), cmap='gray') 
plt.title(f"پیش‌بینی شده: {predicted_class_index}\nاطمینان: {confidence:.4f}")
plt.axis('off') # حذف محورها
plt.show()

print("پیش‌بینی و نمایش تصویر انجام شد.")
```
